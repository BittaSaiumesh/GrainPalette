```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import os

# Load the model
model = load_model("rice_classifier.h5")

# Class labels (must match the training order)
class_labels = ['Arborio', 'Basmati', 'Ipsala', 'Jasmine', 'Karacadag']

# Path to a test image (replace with your image path)
img_path = "../data/Rice_Image_Dataset_split/validation/Basmati/10079.jpg"

# Load and preprocess image
img = image.load_img(img_path, target_size=(64, 64))  # use same size as training
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0  # normalize

# Predict
pred = model.predict(img_array)
predicted_class = class_labels[np.argmax(pred)]

# Display result
plt.imshow(img)
plt.title(f"Predicted: {predicted_class}")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image

# Load the model
model = load_model("rice_classifier.h5")

# Class labels
class_labels = ['Arborio', 'Basmati', 'Ipsala', 'Jasmine', 'Karacadag']

# Image path (use real file)
img_path = "../data/Rice_Image_Dataset_split/validation/Basmati/basmati (10014).jpg"

# Load and preprocess image
img = image.load_img(img_path, target_size=(64, 64))  # match model input size
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0  # normalize

# Predict
pred = model.predict(img_array)
predicted_class = class_labels[np.argmax(pred)]

# Show result
plt.imshow(img)
plt.title(f"Predicted: {predicted_class}")
plt.axis('off')
plt.show()
```
